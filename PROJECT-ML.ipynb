3.Tenure vs Exited:This plot shows the relationship between the length of time customers have been with the bank (tenure) and their likely to be churn.
here we can understand that Majorty of customers have the tenure from 1 to 9

Applying Hyper parameter Tuning

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint as sp_randint

rfc= RandomForestClassifier()
```

```python
param_dist = {"n_estimators": sp_randint(5,200),"max_depth": [5, None],
              "max_features": sp_randint(1, 11),"min_samples_split": sp_randint(2, 11),
              "min_samples_leaf": sp_randint(1, 11),"bootstrap": [True, False],
              "criterion": ["gini", "entropy"]}
```

```python
randomCV = RandomizedSearchCV(rfc,param_distributions=param_dist,cv=10)
randomCV.fit(x_train, y_train)
```
